```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for plots
sns.set(style="whitegrid")

# File paths for the Excel files
forest_file = "Bird_Monitoring_Data_FOREST.xlsx"
grassland_file = "Bird_Monitoring_Data_GRASSLAND.xlsx"

# Function to load all sheets from an Excel file
def load_excel_sheets(file_path):
    """
    Load all sheets from an Excel file into a dictionary of DataFrames.
    """
    excel_file = pd.ExcelFile(file_path)
    sheets = {sheet_name: excel_file.parse(sheet_name) for sheet_name in excel_file.sheet_names}
    return sheets

# Load all sheets from both Excel files
forest_sheets = load_excel_sheets(forest_file)
grassland_sheets = load_excel_sheets(grassland_file)

# Function to perform EDA on a single sheet
# Function to perform EDA on a single sheet
def perform_eda(sheet_name, data):
    """
    Perform EDA on a single sheet.
    """
    print(f"\n--- EDA for Sheet: {sheet_name} ---\n")
    
    # Display the first few rows
    print("First 5 rows of the dataset:")
    display(data.head())
    
    # Number of columns
    print("\nNumber of columns in the dataset:")
    print(len(data.columns))
    
    # Missing value information
    print("\nMissing value information:")
    print(data.isnull().sum())
    
    # Data types
    print("\nData types of each column:")
    print(data.dtypes)
    
    # Statistical details
    print("\nStatistical details of the dataset:")
    print(data.describe())
    
    # Plot the distribution of numerical columns
    print("\nPlot: Distribution of numerical columns")
    numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns
    
    if len(numerical_columns) > 0:
        # Plot histograms only if numerical columns exist
        data[numerical_columns].hist(figsize=(12, 8), bins=20, color='skyblue', edgecolor='black')
        plt.suptitle(f"Distribution of Numerical Columns - {sheet_name}", fontsize=16)
        plt.show()
    else:
        print(f"No numerical columns found in sheet: {sheet_name}. Skipping histogram plot.")


# Perform EDA for each sheet in the forest dataset
print("\n=== EDA for Forest Dataset ===\n")
for sheet_name, data in forest_sheets.items():
    perform_eda(sheet_name, data)

# Perform EDA for each sheet in the grassland dataset
print("\n=== EDA for Grassland Dataset ===\n")
for sheet_name, data in grassland_sheets.items():
    perform_eda(sheet_name, data)

# Combine all sheets into a single DataFrame for each dataset
def combine_sheets(sheets, location_type):
    """
    Combine all sheets into a single DataFrame and add a Location_Type column.
    """
    combined_data = pd.concat(sheets.values(), ignore_index=True)
    combined_data['Location_Type'] = location_type
    return combined_data
```

```python

forest_data = combine_sheets(forest_sheets, "Forest")
grassland_data = combine_sheets(grassland_sheets, "Grassland")

# Combine forest and grassland data into a single DataFrame
combined_data = pd.concat([forest_data, grassland_data], ignore_index=True)
#combined_data.to_csv("cleaned_bird_data.csv", index=False)

# Perform EDA on the combined dataset
print("\n=== EDA for Combined Dataset ===\n")
perform_eda("Combined Dataset", combined_data)
```

```python

# Additional Plots for Combined Data
# Observation Frequency by Year
print("\nPlot: Observation Frequency by Year")
if 'Year' in combined_data.columns:
    yearly_data = combined_data.groupby('Year').size().reset_index(name='Observation_Count')
    plt.figure(figsize=(10, 6))
    sns.barplot(data=yearly_data, x='Year', y='Observation_Count', palette='viridis')
    plt.title("Observation Frequency by Year")
    plt.xlabel("Year")
    plt.ylabel("Number of Observations")
    plt.grid(axis='y')
    plt.show()
```

```python

# Observation Frequency by Month
print("\nPlot: Observation Frequency by Month")
if 'Date' in combined_data.columns:
    combined_data['Month'] = pd.to_datetime(combined_data['Date'], errors='coerce').dt.month
    monthly_data = combined_data.groupby('Month').size().reset_index(name='Observation_Count')
    plt.figure(figsize=(10, 6))
    sns.barplot(data=monthly_data, x='Month', y='Observation_Count', palette='viridis')
    plt.title("Observation Frequency by Month")
    plt.xlabel("Month")
    plt.ylabel("Number of Observations")
    plt.grid(axis='y')
    plt.show()
```

```python

# Biodiversity Hotspots (Top 10 Plots)
print("\nPlot: Top 10 Biodiversity Hotspots (Plots)")
if 'Plot_Name' in combined_data.columns and 'Scientific_Name' in combined_data.columns:
    plot_data = combined_data.groupby('Plot_Name')['Scientific_Name'].nunique().reset_index()
    plot_data = plot_data.sort_values(by='Scientific_Name', ascending=False).head(10)  # Top 10 plots
    plt.figure(figsize=(10, 6))
    sns.barplot(data=plot_data, x='Scientific_Name', y='Plot_Name', palette='viridis')
    plt.title("Top 10 Biodiversity Hotspots (Plots)")
    plt.xlabel("Number of Unique Species")
    plt.ylabel("Plot Name")
    plt.grid(axis='x')
    plt.show()
```

```python

# Top 10 Most Observed Species
print("\nPlot: Top 10 Most Observed Species")
if 'Common_Name' in combined_data.columns:
    species_data = combined_data['Common_Name'].value_counts().head(10).reset_index()
    species_data.columns = ['Common_Name', 'Observation_Count']
    plt.figure(figsize=(10, 6))
    sns.barplot(data=species_data, x='Observation_Count', y='Common_Name', palette='viridis')
    plt.title("Top 10 Most Observed Species")
    plt.xlabel("Number of Observations")
    plt.ylabel("Species")
    plt.grid(axis='x')
    plt.show()
```

```python

# Temperature vs Observation Count
print("\nPlot: Temperature vs Observation Count")
if 'Temperature' in combined_data.columns:
    temp_data = combined_data.groupby('Temperature').size().reset_index(name='Observation_Count')
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=temp_data, x='Temperature', y='Observation_Count', color='blue', alpha=0.7)
    plt.title("Temperature vs Observation Count")
    plt.xlabel("Temperature (Â°C)")
    plt.ylabel("Number of Observations")
    plt.grid()
    plt.show()
```
